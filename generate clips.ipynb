```python
from glob import glob
from pathlib import Path
import pandas as pd
import torch
import librosa
import numpy as np
import matplotlib.pyplot as plt
import random 
from tqdm import tqdm

fnames = glob("ci01/**/**/*flac")
detfnames=glob('detections/ci01/**/data/*flac')
random.shuffle(fnames)
len(fnames)
```

```python
CHUNK = 1_000_000
THRESH = 0.02
peaks=[]
filecount = 0
for fid, f in tqdm(enumerate(fnames)):
    x,sr = librosa.load(f, sr = None)
    for i in range(0, len(x), CHUNK):
        y = x[i:i+CHUNK]
        z=librosa.feature.rms(y=x[i:i+CHUNK])[0]
        for j, zz in enumerate(z):
            if(zz>THRESH):
                peaks.append(i+j)

    for p in peaks:
        clip = x[p-CHUNK//2:p+CHUNK//2]
         
        if(random.random()<0.9):
            np.save(f"clips/train/clip_{fid}_{p}.npy", clip)
        else:
            np.save(f"clips/test/clip_{fid}_{p}.npy", clip)
        filecount+=1
    if(filecount%500==0):
        print(filecount)
    del x
```
