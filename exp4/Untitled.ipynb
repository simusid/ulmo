```python
from mygenerator import MyGenerator
from glob import glob
import matplotlib.pyplot as plt
fnames = glob("/tmp/foo/sancsound/clips/train/*")
```

```python
mygen = MyGenerator(fnames, N=4096)
x,y = mygen.__getitem__(0)
```

```python
signal = x[1]
N=4096
ai = list(range(0,N))
bi = list(range(N,2*N))
ci = list(range(2*N,3*N))
a= signal[0:N]
b = signal[N:2*N]
c = signal[2*N:]
```

```python
plt.figure(figsize=(10,8))
desc = "\nThe model must learn if the Unknown signal is drawn from A or B"
plt.title(f"Self Supervised Pretext Acoustics Task\n{desc}")
plt.plot(ai, a, alpha=.8)
plt.plot(bi, b, alpha=.8)
plt.plot(ci, c, alpha=.8)
plt.text(N-N//2, .85, "Signal A", size='larger')
plt.text(2*N-N//2, .85, "Signal B", size='larger')
plt.text(3*N-N//2-650, .85, "Unknown X", size='larger')
plt.grid()
plt.axis=None
plt.text(4000, -.75, "Note the use of random dropout to make the problem harder",
        bbox= dict(facecolor='w', fill=True, edgecolor='blue', pad=10.0))
plt.savefig("pretext_task.jpg")
```
